```python
def print_matrix(mat, size):
    print("Current Matrix:")
    for i in range (size):
        for j in range (size):
            print(mat[i][j], end = "")
        print()
        
def make_matrix(num_cols,num_rows):
    return [[0 for j in range(num_cols)] for i in range(num_rows)]

def add_Pattern(matrix, inp,desiredOutput, size):
    for i in range (size):
        if inp[i] == "1":
            for j in range (size):
                if desiredOutput[j] == "1" and matrix[i][j] != " ":
                    matrix[i][j] = 1
                    
def retrieve_Pattern(matrix, inp,size):
    threshold = inp.count('1')
    currPattern = ""
    for j in range(size):
        tally = 0
        for i in range(size):
            if inp[i] == "1" and (matrix[i][j] == 1 or matrix[i][j] == " "):
                tally += 1
        if tally >= threshold:
            currPattern+= "1"
        else:
            currPattern += "0"         
    return currPattern

def kill_synapse(matrix, row, col):
    matrix[row][col] = " "
    
```

```python
choice = "";
size = int(input("What is the size of your matrix: "))
mat = make_matrix(size,size)
while choice != "0":
    print()
    print("Choose an option! \n0 to quit. \n1 to add a new pattern \n2 to try and retrieve a pattern \n3 to kill a synapse\n4 to print matrix")
    choice = input()
    if choice == "1":
        inp = input("Choose an Input: ")
        outp = input("Choose a desired output for the pattern: ")
        if len(inp) != size or len(outp) != size :
            print("Input/Desired Output must be the exact size of the matrix. Which is: ",size)
        else:
            add_Pattern(mat,inp,outp,size)
    if choice == "2":
        inp = input("Choose an Input: ")
        if len(inp) > size or len(inp) < size - 1:
            print("Input/Desired Output must be the exact size of the matrix. Which is: ",size)
        else:
            pat = retrieve_Pattern(mat,inp,size)
            print("Your retrieved pattern is: ", pat)
    if choice == "3":
        row = int(input("Choose a row: "))
        col = int(input("Choose a column: "))
        if row > size or col > size or row < 1 or col < 1:
            print("Row/Col cannot be less than 1 (Assume it's 1 indexed) or greater than the size of the matrix. Which is: ", size)
        else:
            kill_synapse(mat,row - 1,col - 1)
    if choice == "4":
        print_matrix(mat,size)
```
